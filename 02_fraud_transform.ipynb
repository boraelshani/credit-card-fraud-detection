```python
df = spark.table("raw_transactions")
df.show(5)
```

```python
df.groupBy("Class").count().show()
```

```python
from pyspark.sql.functions import col

# Check 1: negative or zero amounts (shouldn't exist in real transactions)
invalid_amounts = df.filter(col("Amount") <= 0)
print(f"Invalid amounts (<=0): {invalid_amounts.count()}")

# Check 2: missing/null values in critical columns
null_check = df.filter(col("Amount").isNull() | col("Class").isNull() | col("Time").isNull())
print(f"Rows with missing critical data: {null_check.count()}")

# Check 3: duplicate rows
duplicate_count = df.count() - df.dropDuplicates().count()
print(f"Duplicate rows found: {duplicate_count}")
```

```python
df_clean = df.filter(col("Amount") > 0).dropDuplicates()

print(f"Original rows: {df.count()}")
print(f"Clean rows: {df_clean.count()}")
```

```python
df_clean.write.mode("overwrite").saveAsTable("clean_transactions")
```
